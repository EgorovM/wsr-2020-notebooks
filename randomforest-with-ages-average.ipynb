```python
# Max_Depth < 15
# n_estimators кратные 50, не больше 600

model = RandomForestClassifier(n_estimators=500, max_depth=10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```
