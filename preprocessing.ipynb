```python
import pandas as pd
```

```python
df_train = pd.read_csv('../input/titanic/train.csv')
df_test = pd.read_csv('../input/titanic/test.csv')
```

## Preprocessing

```python
df_train.info()
```

```python
del df_train['Cabin']
```

```python
df_train['Embarked'].value_counts()
```

```python
df_train['Embarked'] = df_train['Embarked'].fillna('S')
```

```python
df_train['Age'] = df_train['Age'].fillna(int(df_train['Age'].mean()))
```

```python
df_train.info()
```

```python
df_train.head()
```

# Обучение модели

```python
# Типы данных

# Вещественные (real)
# Категориальные (categorial)
# Порядковые (categorial)
```

### Прямой способ

```python
# Пол
```

```python
df_train['Sex'].value_counts()
```

```python
df_train.head()
```

```python
is_male = []
is_female = []
```

```python
sexes = df_train['Sex']
```

```python
for i in range(len(sexes)):
    if sexes[i] == 'male':
        is_male.append(1)
        is_female.append(0)
    else:
        is_male.append(0)
        is_female.append(1)
```

```python
df_train['is_male'] = is_male
df_train['is_female'] = is_female
```

```python
# Embarked
```

```python
df_train['Embarked'].value_counts()
```

```python
embs = df_train['Embarked']
```

```python
is_S = []
is_C = []
is_Q = []
```

```python
for i in range(len(embs)):
    if embs[i] == 'S':
        is_S.append(1)
        is_Q.append(0)
        is_C.append(0)
    elif embs[i] == 'C':
        is_S.append(0)
        is_Q.append(0)
        is_C.append(1)
    else:
        is_S.append(0)
        is_Q.append(1)
        is_C.append(0)
```

```python
df_train['emb_is_S'] = is_S
df_train['emb_is_C'] = is_C
df_train['emb_is_Q'] = is_Q
```

```python
names = df_train['Name']

prefixes = []

for i in range(len(names)):
    name = names[i]
    pref = name.split()[1]
    
    if pref != 'Mr.' and pref != 'Miss.' and pref != 'Mrs.' and pref != 'Master.':
        pref = 'Rare'
        
    prefixes.append(pref)

df_train['Prefix'] = prefixes
```
